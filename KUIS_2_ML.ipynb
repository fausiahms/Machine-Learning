```python
# Impor library yang diperlukan
import pandas as pd
from sklearn.linear_model import LinearRegression

# Siapkan data dalam DataFrame
df = pd.DataFrame(asuransi)

# Pisahkan variabel independen dan dependen
x = df[['Payment']]
y = df[['Insured']]

# Inisialisasi model regresi linier
model = LinearRegression()

# Latih model pada data
model.fit(x, y)

# Dapatkan koefisien regresi dan intersep
koefisien = model.coef_
intersep = model.intercept_

# Cetak koefisien dan intersep
print("Koefisien Regresi:", koefisien)
print("Intersep:", intersep)
```

```python
import matplotlib.pyplot as plt
y_pred = model.predict(x)
plt.scatter(x, y, color='blue', label='Observasi')

# Plot garis regresi
plt.plot(x, y_pred, '-r')

# Label dan judul plot
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Regresi Linier Sederhana')

# Menampilkan legenda
plt.legend()

# Menampilkan plot
plt.show()
```
